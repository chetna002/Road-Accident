```python
#Most common causes of accident
```

```python
accident= df.groupby(['Cause category','Outcome of Incident']).max()[['Count']].sort_values('Count',ascending=False)
```

```python
accident
```

```python
accident_reset = accident.reset_index()
```

```python
pivot_table = accident_reset.pivot(index='Cause category', columns='Outcome of Incident', values='Count').fillna(0)
```

```python
plt.figure(figsize=(11, 5))
sns.heatmap(pivot_table, annot=True, fmt='g', cmap='YlGnBu',linecolor='black',linewidths=0.7,alpha=0.9)
plt.gcf().set_facecolor('#2F4F7F')
plt.title('To Identify the most common causes of accidents',fontweight='bold',color='black',fontsize=14,fontstyle='italic')
plt.xlabel("Outcome of Incident",fontweight='bold',fontsize=14)
plt.ylabel('Cause category',fontweight='bold',fontsize=14)

plt.show()
```

1.Road Features account for the highest number of total injuries (3,148) and accidents (3,134). 2.Traffic Violations also significantly contribute to accidents (3,032) and injuries (2,801). 3.Weather conditions are another major factor, with (2,589) injuries and 651 deaths. 4.Minor Injuries are predominantly caused by Road Features (2,676), Traffic Violations (2,423), and Weather (2,164).

Visualization: The heatmap visualizes this data, with darker colors indicating higher values. This makes it easy to spot which causes lead to the most accidents and severe outcomes.

Most Common Causes: Road Features and Traffic Violations stand out with the highest numbers of accidents and injuries. Fatal Incidents: Although fewer, fatalities are highest in Road Features, Traffic Violations, and Weather-related incidents.

```python
def my_graph(a,col):
    data = df[df['Cause category']==f"{a}"]
    data=data.groupby('Outcome of Incident').sum()[['Count']].sort_values('Count',ascending=False)
    
    plt.figure(figsize=(12, 6))
    ax=plt.bar(data.index, data['Count'], edgecolor='black',color=f'{col}')
    plt.title(f'Outcome of Incident on the bases of {a}', fontsize=18,fontweight='bold')
    plt.xlabel('Incidents', fontsize=18,fontweight='bold')
    plt.ylabel('Incident Count', fontsize=14,fontweight='bold')
    plt.xticks(rotation=90)
    sns.set_context("paper")
    sns.set_style("darkgrid")
    plt.bar_label(ax, padding=3,fontweight='bold',size=14) 
        
    plt.tight_layout()
    plt.gcf().set_facecolor('#B3C99C')
    plt.show()
```

Each slice of the pie corresponds to one of the top cities, with the size of the slice representing the proportion of the city's 'Count' relative to the total of the top 10 cities. The largest slice signifies the city with the highest count, visually emphasized by an exploded slice to highlight the most significant contributor.

Chennai stands out as the leading city, with 78,459 incidents, making it the highest contributor among the cities. Delhi follows with 72,526 incidents, while Jabalpur records 57,715 incidents. The chart provides a clear comparison of incident numbers across these cities, allowing for easy identification of the highest and lowest counts within the top 10.
